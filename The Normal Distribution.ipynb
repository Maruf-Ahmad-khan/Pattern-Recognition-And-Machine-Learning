```python
import pandas as pd
import numpy as np
from scipy.stats import norm
class Solution:
     
     def __init__(self, mean, std_dev):
          self.mean = mean
          self.std_dev = std_dev
          
     def z_score(self, value):
          """
          Calculate  the z-score of a value
          """
          return (value - self.mean) / self.std_dev
     
     
     def Cumulative_Probability(self, value):
          """"
          Calculate  the cumulative probability of a value in a normal distribution
          """
          z = self.z_score(value)
          return norm.cdf(z)
     
     
     def Probability(self, value):
          """"
          Calculate the probability of a specific values
          """
          return self.Cumulative_Probability(value)
     
"""
Read the data  from the CSV file
"""
data = pd.read_csv('Test set.csv')

"""
Calculate mean, std for Height column
"""
mean = int(np.ceil(data['Height'].mean()))
std_dev = int(np.ceil(data['Height'].std()))

"""
Create an instance for Solution class
"""
sol = Solution(mean, std_dev)

"""
Apply z-score function to each values of Height column
"""
data['z_score_height'] =  data['Height'].apply(sol.z_score).round(2)

"""
Apply the Cummulative_Probability function to  each values of Height column
"""

data['Cumulative_Probability_height'] = data['Height'].apply(sol.Cumulative_Probability).round(2)

"""
Apply z-score function to each values of Weight column
"""
data['z_score_weight'] =  data['Weight'].apply(sol.z_score).round(2)

"""
Apply the Cummulative_Probability function to  each values of Weight column
"""
data['Cumulative_Probability_weight'] = data['Weight'].apply(sol.Cumulative_Probability)

data.head(10)
          
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

class StudentsPerformance:
    def __init__(self, mean, std_dev):
        """Initialize the StudentsPerformance class with mean and standard deviation."""
        self.mean = mean
        self.std_dev = std_dev

    def z_score(self, value):
        """Calculate the Z-score for a given value."""
        if self.std_dev == 0:
            raise ValueError("Standard deviation cannot be zero.")
        return (value - self.mean) / self.std_dev

    def cumulative_probability(self, value):
        """Calculate the cumulative probability (CDF) for a given value."""
        z = self.z_score(value)
        return norm.cdf(z)

    def probability(self, value):
        """Calculate the probability for a specific value."""
        return self.cumulative_probability(value)


# Load the data
df = pd.read_csv(r"C:\Users\mk744\Downloads\StudentPerformanceFactors.csv")

# Calculate mean and standard deviation for 'Hours_Studied' column
mean = df['Hours_Studied'].mean()
std_dev = df['Hours_Studied'].std()

# Create an instance of the StudentsPerformance class
sol = StudentsPerformance(mean, std_dev)

# Apply the z_score and cumulative_probability functions to the 'Hours_Studied' column
df['z_score_hours_studied'] = sol.z_score(df['Hours_Studied']).round(2)
df['probability_hours_studied'] = sol.cumulative_probability(df['Hours_Studied']).round(2)

# Filter to show relevant columns
df_filtered = df[['Hours_Studied', 'z_score_hours_studied', 'probability_hours_studied']]

# Display the first few rows
print(df_filtered.head())
```
