```python
import pandas as pd
import numpy as np
from scipy.stats import norm
class Solution:
     
     def __init__(self, mean, std_dev):
          self.mean = mean
          self.std_dev = std_dev
          
     def z_score(self, value):
          """
          Calculate  the z-score of a value
          """
          return (value - self.mean) / self.std_dev
     
     
     def Cumulative_Probability(self, value):
          """"
          Calculate  the cumulative probability of a value in a normal distribution
          """
          z = self.z_score(value)
          return norm.cdf(z)
     
     
     def Probability(self, value):
          """"
          Calculate the probability of a specific values
          """
          return self.Cumulative_Probability(value)
     
"""
Read the data  from the CSV file
"""
data = pd.read_csv('Test set.csv')

"""
Calculate mean, std for Height column
"""
mean = int(np.ceil(data['Height'].mean()))
std_dev = int(np.ceil(data['Height'].std()))

"""
Create an instance for Solution class
"""
sol = Solution(mean, std_dev)

"""
Apply z-score function to each values of Height column
"""
data['z_score_height'] =  data['Height'].apply(sol.z_score).round(2)

"""
Apply the Cummulative_Probability function to  each values of Height column
"""

data['Cumulative_Probability_height'] = data['Height'].apply(sol.Cumulative_Probability).round(2)

"""
Apply z-score function to each values of Weight column
"""
data['z_score_weight'] =  data['Weight'].apply(sol.z_score).round(2)

"""
Apply the Cummulative_Probability function to  each values of Weight column
"""
data['Cumulative_Probability_weight'] = data['Weight'].apply(sol.Cumulative_Probability)

data.head(10)
          
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

class StudentsPerformance:
    def __init__(self, mean, std_dev):
        """Initialize the StudentsPerformance class with mean and standard deviation."""
        self.mean = mean
        self.std_dev = std_dev

    def z_score(self, value):
        """Calculate the Z-score for a given value."""
        if self.std_dev == 0:
            raise ValueError("Standard deviation cannot be zero.")
        return (value - self.mean) / self.std_dev

    def cumulative_probability(self, value):
        """Calculate the cumulative probability (CDF) for a given value."""
        z = self.z_score(value)
        return norm.cdf(z)

    def probability(self, value):
        """Calculate the probability for a specific value."""
        return self.cumulative_probability(value)


# Load the data
df = pd.read_csv(r"C:\Users\mk744\Downloads\StudentPerformanceFactors.csv")

# Calculate mean and standard deviation for 'Hours_Studied' column
mean = df['Hours_Studied'].mean()
std_dev = df['Hours_Studied'].std()

# Create an instance of the StudentsPerformance class
sol = StudentsPerformance(mean, std_dev)

# Apply the z_score and cumulative_probability functions to the 'Hours_Studied' column
df['z_score_hours_studied'] = sol.z_score(df['Hours_Studied']).round(2)
df['probability_hours_studied'] = sol.cumulative_probability(df['Hours_Studied']).round(2)

# Filter to show relevant columns
df_filtered = df[['Hours_Studied', 'z_score_hours_studied', 'probability_hours_studied']]

# Display the first few rows
print(df_filtered.head())
```

# Using Shapiro-test

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

class NormalityCheck:
    """
    A class to perform normality checks and visualizations on a dataset.
    """
    
    def __init__(self, dataset):
        """
        Initialize the class with the dataset.
        Args:
            dataset (pd.DataFrame): The dataset containing numerical columns to check for normality.
        """
        self.dataset = dataset
        self.numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns
        self.normality_results = {}
    
    def check_normality(self):
        """
        Perform the Shapiro-Wilk test for normality on all numerical columns.
        
        Returns:
            dict: A dictionary containing the test statistics and p-values for each numerical column.
        """
        for column in self.numerical_columns:
            try:
                stat, p_value = stats.shapiro(self.dataset[column].dropna())  # Drop NA to avoid errors
                self.normality_results[column] = {
                    'Shapiro-Wilk Statistic': stat,
                    'p-value': p_value,
                    'normal_distribution': p_value > 0.05  # True if likely normal
                }
            except ValueError as e:
                print(f"Error processing column {column}: {e}")
                self.normality_results[column] = {'Error': str(e)}
        
        return self.normality_results
    
    def visualize_distribution(self, columns=None, save_plots=False, plot_dir='plots'):
        """
        Visualize the distribution of the specified numerical columns using histograms and QQ plots.
        
        Args:
            columns (list, optional): List of columns to plot. If None, all numerical columns are plotted.
            save_plots (bool, optional): Whether to save the plots to files instead of showing them.
            plot_dir (str, optional): Directory to save the plots if save_plots is True.
        """
        if columns is None:
            columns = self.numerical_columns
        
        for column in columns:
            if column in self.dataset.columns:
                plt.figure(figsize=(12, 6))
                
                # Histogram with KDE
                plt.subplot(1, 2, 1)
                sns.histplot(self.dataset[column], kde=True, bins=20, edgecolor='black')
                plt.title(f'Histogram of {column}')
                
                # QQ Plot
                plt.subplot(1, 2, 2)
                stats.probplot(self.dataset[column], dist="norm", plot=plt)
                plt.title(f'QQ Plot of {column}')
                
                plt.tight_layout()
                
                if save_plots:
                    plt.savefig(f"{plot_dir}/{column}_distribution.png")
                else:
                    plt.show()
    
    def print_normality_results(self):
        """
        Print the results of the Shapiro-Wilk normality test for each column.
        """
        for column, result in self.normality_results.items():
            if 'Error' not in result:
                normal_status = "likely follows a normal distribution" if result['normal_distribution'] else "does NOT follow a normal distribution"
                print(f"'{column}' {normal_status} (p-value: {result['p-value']:.4f})")
            else:
                print(f"Error with column {column}: {result['Error']}")
    
    def save_results(self, filename='normality_results.csv'):
        """
        Save the normality test results to a CSV file.
        
        Args:
            filename (str): The name of the file to save the results to.
        """
        df_results = pd.DataFrame(self.normality_results).T
        df_results.to_csv(filename)
        print(f"Results saved to {filename}")

if __name__ == "__main__":
    # Load the dataset
    data = pd.read_csv(r"C:\Users\mk744\Downloads\StudentPerformanceFactors.csv")

    # Create an instance of the class
    normality_checker = NormalityCheck(data)

    # Check normality for all numerical columns
    results = normality_checker.check_normality()

    # Print the results of the Shapiro-Wilk normality test
    normality_checker.print_normality_results()

    # Visualize the distributions of the numerical columns
    normality_checker.visualize_distribution(save_plots=False)

    # Save the results to a CSV file
    normality_checker.save_results('normality_results.csv')
```

# Using Andersion Darling-Test

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import pandas as pd

class NormalityCheckAD:
    """
    A class to perform normality checks using the Anderson-Darling test and visualizations on a dataset.
    """
    
    def __init__(self, dataset):
        """
        Initialize the class with the dataset.
        Args:
            dataset (pd.DataFrame): The dataset containing numerical columns to check for normality.
        """
        self.dataset = dataset
        self.numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns
        self.normality_results = {}
    
    def check_normality(self):
        """
        Perform the Anderson-Darling test for normality on all numerical columns.
        
        Returns:
            dict: A dictionary containing the test statistic and critical values for each numerical column.
        """
        for column in self.numerical_columns:
            try:
                result = stats.anderson(self.dataset[column].dropna(), dist='norm')
                self.normality_results[column] = {
                    'AD Statistic': result.statistic,
                    'Critical Values': result.critical_values,
                    'Significance Levels': result.significance_level,
                    'normal_distribution': result.statistic < result.critical_values[2]  # At 5% significance level
                }
            except ValueError as e:
                print(f"Error processing column {column}: {e}")
                self.normality_results[column] = {'Error': str(e)}
        
        return self.normality_results
    
    def visualize_distribution(self, columns=None, save_plots=False, plot_dir='plots'):
        """
        Visualize the distribution of the specified numerical columns using histograms and QQ plots.
        """
        if columns is None:
            columns = self.numerical_columns
        
        for column in columns:
            if column in self.dataset.columns:
                plt.figure(figsize=(12, 6))
                
                # Histogram with KDE
                plt.subplot(1, 2, 1)
                sns.histplot(self.dataset[column], kde=True, bins=20, edgecolor='black')
                plt.title(f'Histogram of {column}')
                
                # QQ Plot
                plt.subplot(1, 2, 2)
                stats.probplot(self.dataset[column], dist="norm", plot=plt)
                plt.title(f'QQ Plot of {column}')
                
                plt.tight_layout()
                
                if save_plots:
                    plt.savefig(f"{plot_dir}/{column}_distribution.png")
                else:
                    plt.show()
    
    def print_normality_results(self):
        """
        Print the results of the Anderson-Darling normality test for each column.
        """
        for column, result in self.normality_results.items():
            if 'Error' not in result:
                normal_status = "likely follows a normal distribution" if result['normal_distribution'] else "does NOT follow a normal distribution"
                print(f"'{column}' {normal_status} (AD Statistic: {result['AD Statistic']:.4f})")
            else:
                print(f"Error with column {column}: {result['Error']}")
    
    def save_results(self, filename='normality_results_ad.csv'):
        """
        Save the normality test results to a CSV file.
        """
        df_results = pd.DataFrame(self.normality_results).T
        df_results.to_csv(filename)
        print(f"Results saved to {filename}")

# Example Usage:
if __name__ == "__main__":
    # Load the dataset
    data = pd.read_csv(r"C:\Users\mk744\Downloads\StudentPerformanceFactors.csv")

    # Create an instance of the class
    normality_checker_ad = NormalityCheckAD(data)

    # Check normality for all numerical columns
    results_ad = normality_checker_ad.check_normality()

    # Print the results of the Anderson-Darling normality test
    normality_checker_ad.print_normality_results()

    # Visualize the distributions of the numerical columns
    normality_checker_ad.visualize_distribution(save_plots=False)

    # Save the results to a CSV file
    normality_checker_ad.save_results('normality_results_ad.csv')
```

# Using Jarque-Bera Test:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import pandas as pd

class NormalityCheckJB:
    """
    A class to perform normality checks using the Jarque-Bera test and visualizations on a dataset.
    """
    
    def __init__(self, dataset):
        """
        Initialize the class with the dataset.
        Args:
            dataset (pd.DataFrame): The dataset containing numerical columns to check for normality.
        """
        self.dataset = dataset
        self.numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns
        self.normality_results = {}
    
    def check_normality(self):
        """
        Perform the Jarque-Bera test for normality on all numerical columns.
        
        Returns:
            dict: A dictionary containing the test statistic and p-values for each numerical column.
        """
        for column in self.numerical_columns:
            try:
                jb_stat, p_value = stats.jarque_bera(self.dataset[column].dropna())
                self.normality_results[column] = {
                    'JB Statistic': jb_stat,
                    'p-value': p_value,
                    'normal_distribution': p_value > 0.05  # True if likely normal
                }
            except ValueError as e:
                print(f"Error processing column {column}: {e}")
                self.normality_results[column] = {'Error': str(e)}
        
        return self.normality_results
    
    def visualize_distribution(self, columns=None, save_plots=False, plot_dir='plots'):
        """
        Visualize the distribution of the specified numerical columns using histograms and QQ plots.
        """
        if columns is None:
            columns = self.numerical_columns
        
        for column in columns:
            if column in self.dataset.columns:
                plt.figure(figsize=(12, 6))
                
                # Histogram with KDE
                plt.subplot(1, 2, 1)
                sns.histplot(self.dataset[column], kde=True, bins=20, edgecolor='black')
                plt.title(f'Histogram of {column}')
                
                # QQ Plot
                plt.subplot(1, 2, 2)
                stats.probplot(self.dataset[column], dist="norm", plot=plt)
                plt.title(f'QQ Plot of {column}')
                
                plt.tight_layout()
                
                if save_plots:
                    plt.savefig(f"{plot_dir}/{column}_distribution.png")
                else:
                    plt.show()
    
    def print_normality_results(self):
        """
        Print the results of the Jarque-Bera normality test for each column.
        """
        for column, result in self.normality_results.items():
            if 'Error' not in result:
                normal_status = "likely follows a normal distribution" if result['normal_distribution'] else "does NOT follow a normal distribution"
                print(f"'{column}' {normal_status} (p-value: {result['p-value']:.4f})")
            else:
                print(f"Error with column {column}: {result['Error']}")
    
    def save_results(self, filename='normality_results_jb.csv'):
        """
        Save the normality test results to a CSV file.
        """
        df_results = pd.DataFrame(self.normality_results).T
        df_results.to_csv(filename)
        print(f"Results saved to {filename}")


if __name__ == "__main__":
    # Load the dataset
    data = pd.read_csv(r"C:\Users\mk744\Downloads\StudentPerformanceFactors.csv")

    # Create an instance of the class
    normality_checker_jb = NormalityCheckJB(data)

    # Check normality for all numerical columns
    results_jb = normality_checker_jb.check_normality()

    # Print the results of the Jarque-Bera normality test
    normality_checker_jb.print_normality_results()

    # Visualize the distributions of the numerical columns
    normality_checker_jb.visualize_distribution(save_plots=False)

    # Save the results to a CSV file
    normality_checker_jb.save_results('normality_results_jb.csv')
```
