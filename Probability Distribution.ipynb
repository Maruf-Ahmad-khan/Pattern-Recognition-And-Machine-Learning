# Use Case of Probability Distributions
-To understand how distributions and the probability of events work, let's consider a real-life example involving the insurance industry. -Insurance companies rely heavily on probability distributions to assess risk and set premiums.

- Real-Life Example: Car Insurance
- Insurance companies use probability distributions to predict the likelihood of an event such as a car accident. Based on historical data, - they can determine the probability of accidents for different age groups, driving histories, and vehicle types. This information helps - -- them calculate premiums and manage risk.

- Scenario:
- Age Group Distribution: Younger drivers (18-25) have a higher probability of accidents compared to older drivers (over 25).
- Driving History: Drivers with previous accidents have a higher likelihood of future accidents.
- Vehicle Type: Sports cars have a higher risk of accidents compared to sedans.

# Driver Class

```python
class Driver:
    def __init__(self, age, has_previous_accidents, vehicle_type):
        self.age = age
        self.has_previous_accidents = has_previous_accidents
        self.vehicle_type = vehicle_type

    def calculate_risk_score(self):
        risk_score = 0
        # Age factor
        if self.age <= 25:
            risk_score += 3
        else:
            risk_score += 1
        
        # Previous accidents factor
        if self.has_previous_accidents:
            risk_score += 5
        else:
            risk_score += 1
        
        # Vehicle type factor
        if self.vehicle_type == 'sports':
            risk_score += 4
        else:
            risk_score += 1
        
        return risk_score
```

# Create List Of Driver

```python
drivers = [
    Driver(20, True, 'sports'),
    Driver(30, False, 'sedan'),
    Driver(22, False, 'sports'),
    Driver(35, True, 'sedan'),
    Driver(28, False, 'sedan'),
    Driver(19, True, 'sports'),
]
```

# Calculate Risk Score

```python
risk_scores = [driver.calculate_risk_score() for driver in drivers]
print(risk_scores)
```

# Visualization of Data

```python
import matplotlib.pyplot as plt

# Separate the data for visualization
ages = [driver.age for driver in drivers]
previous_accidents = [driver.has_previous_accidents for driver in drivers]
vehicle_types = [driver.vehicle_type for driver in drivers]

# Visualization
plt.figure(figsize=(10, 6))

# Age vs. Risk Score
plt.subplot(2, 2, 1)
plt.scatter(ages, risk_scores, color='blue')
plt.title('Age vs. Risk Score', color = 'white')
plt.xlabel('Age', color = 'white')
plt.ylabel('Risk Score')
plt.xticks(color = 'white')
plt.yticks(color = 'white')


# Previous Accidents vs. Risk Score
plt.subplot(2, 2, 2)
plt.scatter(previous_accidents, risk_scores, color='red')
plt.title('Previous Accidents vs. Risk Score', color = 'white')
plt.xlabel('Previous Accidents (0: No, 1: Yes)', color = 'white')
plt.ylabel('Risk Score', color = 'white')
plt.xticks(color = 'white')
plt.yticks(color = 'white')


# Vehicle Type vs. Risk Score
plt.subplot(2, 2, 3)
vehicle_numeric = [1 if vt == 'sports' else 0 for vt in vehicle_types]
plt.scatter(vehicle_numeric, risk_scores, color='green')
plt.title('Vehicle Type vs. Risk Score', color = 'white')
plt.xlabel('Vehicle Type (0: Sedan, 1: Sports)', color = 'white')
plt.ylabel('Risk Score', color = 'white')
plt.xticks(color = 'white')
plt.yticks(color = 'white')

plt.tight_layout()
plt.show()
```

# Concept Explanation
Probability Density Function (PDF) 𝑝(𝑥)
p(x)

The PDF describes the likelihood of a random variable 
𝑥
x taking on a specific value.
The area under the PDF curve over an interval represents the probability that the variable falls within that interval.



Cumulative Distribution Function (CDF) 𝑃(𝑥)
P(x)
The CDF gives the probability that a random variable 
𝑥
x will be less than or equal to a certain value.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class Relationship:
    def __init__(self, x_range, mean, std):
        self.x = np.linspace(x_range[0], x_range[1], 1000)
        self.mean = mean
        self.std = std
        self.pdf = self.calculate_pdf(self.x, self.mean, self.std)
        self.cdf = self.calculate_cdf(self.pdf)
    
    def calculate_pdf(self, x, mean, std):
        return (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)
    
    def calculate_cdf(self, pdf):
        return np.cumsum(pdf) * (self.x[1] - self.x[0])
    
    def plot(self):
        plt.figure(figsize=(10, 6))
        sns.set(style="darkgrid")
        
        fig, ax1 = plt.subplots()

        ax1.plot(self.x, self.pdf, 'r-', label='PDF (p(x))')
        ax1.set_xlabel('x')
        ax1.set_ylabel('PDF (p(x))', color='r')
        ax1.tick_params('y', colors='r')

        ax2 = ax1.twinx()
        ax2.plot(self.x, self.cdf, 'b-', label='CDF (P(x))')
        ax2.set_ylabel('CDF (P(x))', color='b')
        ax2.tick_params('y', colors='b')
        
        ax1.legend(loc='upper left')
        ax2.legend(loc='upper right')

        plt.title('Relationship between PDF and CDF')
        plt.show()

# Instantiate and plot the relationship
relationship = Relationship(x_range=(-5, 5), mean=0, std=1)
relationship.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data points
data_points = np.array([1, 2, 2.5, 3, 4, 5, 5.5, 6, 7, 8])

# Parameters for the Gaussian distribution
mu = np.mean(data_points)
sigma = np.std(data_points)

# Creating a range of x values for the Gaussian curve
x = np.linspace(min(data_points) - 1, max(data_points) + 1, 100)

# Gaussian distribution curve
gaussian_curve = norm.pdf(x, mu, sigma)

# Plotting the data points and the Gaussian distribution
plt.figure(figsize=(10, 6))
plt.plot(x, gaussian_curve, 'r-', label=f'Gaussian Distribution\n$\\mu={mu:.2f}$, $\\sigma={sigma:.2f}$')
plt.scatter(data_points, norm.pdf(data_points, mu, sigma), color='blue', zorder=5)
plt.vlines(data_points, ymin=0, ymax=norm.pdf(data_points, mu, sigma), colors='green', linestyles='dotted', lw=1)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Likelihood Function for a Gaussian Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=5, scale=2, size=100)

# Calculate MLE estimates
mu_hat = np.mean(sample_data)
sigma2_hat = np.var(sample_data)

# Plot the data and the estimated Gaussian distribution
x = np.linspace(min(sample_data) - 1, max(sample_data) + 1, 1000)
pdf = norm.pdf(x, mu_hat, np.sqrt(sigma2_hat))

plt.figure(figsize=(10, 6))
plt.hist(sample_data, bins=20, density=True, alpha=0.6, color='g', label='Sample Data Histogram')
plt.plot(x, pdf, 'r-', lw=2, label=f'Estimated Gaussian\n$\mu={mu_hat:.2f}$, $\sigma^2={sigma2_hat:.2f}$')
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.title('Sample Data and Estimated Gaussian Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

class Solution:
    def __init__(self, data, mean, std_dev):
        self.data = data
        self.mean = mean
        self.std_dev = std_dev

    def z_score(self, value):
        # Method to calculate Z-score for a specific value
        return (value - self.mean) / self.std_dev

    def cumulative_probability(self, value):
        # Method to calculate cumulative probability (CDF) for a specific value
        z = self.z_score(value)
        return norm.cdf(z)

    def probability(self, value):
        # Method to calculate the probability for a specific value
        return self.cumulative_probability(value)

# Load the dataset
data = pd.read_csv('Test set.csv')

# Calculate mean and standard deviation for the 'Weight' column
mean = int(np.ceil(data['Weight'].mean()))
std_dev = int(np.ceil(data['Weight'].std()))

# Create an instance of the Solution class
sol = Solution(data, mean, std_dev)

# Apply the z_score function to each value in the 'Weight' column
data['Z_Score'] = data['Weight'].apply(sol.z_score)

# Apply the cumulative_probability function to each value in the 'Weight' column
data['Cumulative_Probability'] = data['Weight'].apply(sol.cumulative_probability)
# Display the updated DataFrame
print(data.head())
```
