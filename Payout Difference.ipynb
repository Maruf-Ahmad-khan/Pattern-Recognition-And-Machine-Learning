```python
class QuickSort:
    
    def swap(self, arr, i, j):
        """
        Swap the elements at index i and j in the array.
        """
        temp = arr[i]
        arr[i] = arr[j]
        arr[j] = temp

    def partition(self, arr, start, end):
        """
        Partition the array around a pivot element.
        Pivot is selected as the first element (arr[start]).
        After partition, all elements smaller than the pivot will be on the left,
        and all elements greater than the pivot will be on the right.
        """
        pivot = arr[start]  # choose the first element as the pivot
        i = start  # pointer for the boundary of elements less than the pivot
        
        for j in range(start + 1, end + 1):
            if arr[j] <= pivot:
                i += 1  # move boundary to the right
                self.swap(arr, i, j)  # swap element at i with element at j
        
        self.swap(arr, i, start)  # place the pivot in the correct position
        return i  # return the index of the pivot

    def quicksort(self, arr, start, end):
        """
        The main recursive function to sort the array using the Quicksort algorithm.
        It recursively sorts the left and right partitions around the pivot element.
        """
        if start >= end:
            return  # base case: if the array has one or zero elements
        
        # Partition the array and get the pivot's index
        pivot_index = self.partition(arr, start, end)
        
        # Recursively sort elements before the pivot and after the pivot
        self.quicksort(arr, start, pivot_index - 1)  # left partition
        self.quicksort(arr, pivot_index + 1, end)    # right partition

    def print_array(self, arr):
        """
        Utility function to print the array.
        """
        print(" ".join(map(str, arr)))


# Example usage:
if __name__ == "__main__":
    arr = [5, 3, 8, 4, 2]  # example array to sort
    sorter = QuickSort()    # create an instance of the QuickSort class
    
    print("Original array:")
    sorter.print_array(arr)  # print original array
    
    sorter.quicksort(arr, 0, len(arr) - 1)  # call quicksort to sort the array
    
    print("Sorted array:")
    sorter.print_array(arr)  # print the sorted array
```
