```python
import pandas as pd

# Load the Titanic dataset directly from a URL (example URL from a public repo)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)

# Display the first few rows
titanic.head()
```

```python
titanic.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(strategy='most_frequent')
df_Imputed = imputer.fit_transform(titanic)
data = pd.DataFrame(df_Imputed, columns=titanic.columns)
data.head()
```

```python
data.isna().sum()
```

```python
data.columns.tolist()
```

```python
Ans = data.filter(items=[
                   'PassengerId', 'Survived', 'Pclass',
                   'Age','Fare', 'SibSp', 'Parch'
                   ])
Ans
```

```python
Ans.isna().sum()
```

```python
Ans.query("Age == 22")
```

```python
for (Age, group) in Ans.groupby('Age'):
     print("{0:30} shape={1}".format(Age, group.shape))
```

```python
# Group data by a specific column (in this case, Age) to perform further analysis.
for age, group in Ans.groupby('Age'):
    print(f"{age:<30} shape={group.shape}")
```
